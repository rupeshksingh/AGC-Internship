````python
def analysis(result_path):

    catch_result = pd.read_csv(result_path)

    print("=" * 80)
    print(catch_result.head())

    print("=" * 80)
    print(f"Number of anomalous points : {catch_result['True Label'].sum()}")

    print("=" * 80)
    fig = px.scatter(catch_result, x=catch_result.index, y="Speed", color="Predicted Label", symbol="True Label")
    fig.show()

    print("=" * 80)
    count = 0
    ranges = []
    for i, val in enumerate(catch_result["Predicted Label"]):
        if val == 1 and count == 0:
            count = 1
            start = i
        elif val == 0 and count == 1:
            count = 0
            end = i-1
            ranges.append({"start" : start, "end" : end})

    print(f"{len(ranges)} Windows of anomaly Found")
    print("=" * 80)
    
    for r in ranges:
        print(r)

    print("=" * 80)
````

````python
analysis("Plots & Results\catch_result.csv")
````
